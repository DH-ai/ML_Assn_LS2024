```python
# Use the file namd 'training data' to train the model

data = pd.read_excel('Training data.xlsx')
x_train = np.array(data.iloc[:,0:8])
y_train = np.array(data.iloc[:,8]).reshape(-1,1)
print(data.columns)
# print(x_train)

# print(data.head(0).columns[1])
# Try plotting y_train with different features
# To get an idea whether to add some features or not
# Add some features if required in x_train

# n = x_train.shape[1]
# for i in range(n):
#   plt.scatter(x_train[:,i],y_train)
#   plt.xlabel(data.head(0).columns[i])
#   plt.ylabel('marks')
#   plt.show()

print(data.isna().sum())
# print(data['iq'].unique())

# Also do label encoding for features not represented in numbers
# refer the link if not know : https://youtu.be/589nCGeWG1w?si=t2Wa7LgbUOO4RooM

def feature_changing(x_train,df):
  # ---------
    # Your code here
  
  df2 = df.drop(columns=['sex','age','internet'])
  x_train = np.array(df2.iloc[:,0:8])
  return x_train

  
  # ---------

x_train = feature_changing(x_train,data)
print(x_train[0])
```

```python
def cost(x_train,y_train,w,b):

  m = x_train.shape[0]
  loss =0
  for i in range(m):
    loss+=((np.dot(x_train[i],w)+b)-y_train[i])**2
  loss/=2*m
  
  return loss[0]
```

```python
def gradient(X,Y,w,b):
    m,n = X.shape # (number of examples, number of features)
    dj_dw = np.zeros ((n,))
    dj_db =0
    for i in range (m):
      err = (np.dot(X[i],w)+b)-Y[i]
      for j in range(n):
         
         dj_dw[j]+=np.dot(X[i,j],err)[0]
         
      dj_db += err
    dj_dw/=m
    dj_db/=m
    return dj_db, dj_dw


def gradient_descent(x_train,y_train,w,b):

  # ---------
    # Your code here
    # Choose learning rate yourself
  # ---------
  alpha = 0.0001
  dj_db, dj_dw = gradient(x_train,y_train,w,b)
  w = w-alpha*dj_dw
  b = b-alpha*dj_db
  return w,b
```

```python
def z_score(data):
    mean = np.mean(data, axis=0)
    std = np.std(data, axis=0)
    standardized_data = (data - mean) / std
    return standardized_data, std, mean
```

```python
x_train = x_train.astype(np.float64)
x_train,x_std,x_mean = z_score(x_train)

np.random.seed(2147483647)
w = np.random.randn(x_train.shape[1],1)
b = np.random.randn(1)

old_cost = 0

while abs(old_cost - float(cost(x_train,y_train,w,b)))>0.00001:
  t = old_cost
  old_cost = cost(x_train,y_train,w,b)
  print("old_cost:",old_cost-t)
  w,b = gradient_descent(x_train,y_train,w,b)


  
df = pd.read_excel('./Test data.xlsx')
x_predict = df.iloc[:,:8].to_numpy()
x_predict = feature_changing(x_predict,df)
x_predict = (x_predict - x_mean)/x_std
ans = df.iloc[:,8].to_numpy()

y_predict = np.dot(x_predict,w) + b

accuracy = 0
for dim in range(len(ans)):
  # print(y_predict[dim,0]-ans[dim])
  if abs(y_predict[dim,0]-ans[dim])<0.5: # do not change the tolerance as you'll be checked on +- 0.5 error only
    accuracy += 1
accuracy = round(accuracy*100/200.0,2)
ok = 'Congratulations' if accuracy>95 else 'Optimization required'
print(f"{ok}, your accuracy is {accuracy}%")
```
